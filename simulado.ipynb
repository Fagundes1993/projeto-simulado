```python
import pandas as pd

def realizar_simulado(df_questoes):
    """
    Permite ao usuário escolher um tema e realizar o simulado correspondente.

    Args:
        df_questoes: DataFrame com as questões do simulado.

    Returns:
        None
    """

    print("Temas disponíveis:")
    for tema in df_questoes['tema'].unique():
        print(f"- {tema}")

    tema_escolhido = input("Escolha um tema: ")

    # Filtrar as questões do tema escolhido
    questoes_tema = df_questoes[df_questoes['tema'] == tema_escolhido]

    # Embaralhar as questões
    questoes_tema = questoes_tema.sample(frac=1)

    acertos = 0
    total_questoes = len(questoes_tema)

    # Apresentar as questões e receber as respostas do usuário
    for i, row in questoes_tema.iterrows():
        print(f"\nQuestão {i+1}:")
        print(row['enunciado'])
        print(f"a) {row['alternativa_a']}")
        print(f"b) {row['alternativa_b']}")
        print(f"c) {row['alternativa_c']}")
        print(f"d) {row['alternativa_d']}")
        print(f"e) {row['alternativa_e']}")

        resposta = input("Sua resposta: ").upper()

        if resposta == row['gabarito']:
            acertos += 1
            print("Resposta correta!")
        else:
            print(f"Resposta incorreta. O gabarito é {row['gabarito']}.")

    # Calcular a porcentagem de acerto
    porcentagem_acerto = (acertos / total_questoes) * 100

    print(f"\nVocê acertou {acertos} de {total_questoes} questões.")
    print(f"Sua porcentagem de acerto: {porcentagem_acerto:.2f}%")

    # Feedback sobre o desempenho
    if porcentagem_acerto >= 90:
        print("Excelente desempenho! Parabéns!")
    elif porcentagem_acerto >= 70:
        print("Bom desempenho! Continue estudando para melhorar.")
    elif porcentagem_acerto >= 50:
        print("Você está na média. É preciso se dedicar mais aos estudos.")
    else:
        print("Seu desempenho precisa melhorar. Revise os conceitos com atenção.")


# Carregar o DataFrame com as questões
df_questoes = pd.read_csv('questoes_simulado.csv')

# Iniciar o simulado
realizar_simulado(df_questoes)
```
